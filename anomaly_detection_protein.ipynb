sampling

```python
df=data.sample(frac=0.95)
unseen =data.drop(df.index)
```

```python
print(df.shape,unseen.shape)
```

```python
df.reset_index(drop=True,inplace=True)
unseen.reset_index(drop=True,inplace=True)
```

```python
df.head()
```

pycaret preprocessing

```python
from pycaret.anomaly import *
```

```python
exp=setup(df,normalize=True,ignore_features=['MouseID'])
```

model creation

```python
models()
```

```python
iforest=create_model('iforest')
```

```python
print(iforest)
```

```python
iforest_result=assign_model(iforest)
```

```python
iforest_result.head()
```

```python
anomaly=iforest_result[iforest_result['Anomaly']==1]
```

```python
anomaly.MouseID
```

```python
plot_model(iforest)
```

```python
plot_model(iforest,plot='umap')
```

prediction on unseen

```python
unseen_pred=predict_model(iforest,data=unseen)
```
